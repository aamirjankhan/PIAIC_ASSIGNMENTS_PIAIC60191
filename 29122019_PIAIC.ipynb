# arguments with dictionaries

```python
def id_card(sID,sName,fName,course,timing):
    text='''
    PIAIC Islamabad
    ID : {}
    Name: {}
    Father's name : {}
    course : {}
    timing : {}
    '''.format(sID,sName,fName, course, timing)
    return text
d1={
    'course':'A.I',
    'sName':'Aamir Jan Khan',
    'fName':'Muhammad Khalid Raheel Khan',
    'sID':'200',
    'timing':'9-12'
}
print(id_card(**d1))
```

# Recursive method

```python
def abc(n):
    if n==1:
        return 1
    return n * abc(n-1)
abc(6)
```

# Generator function

```python
def abc():
    for i in range(11):
        yield i
a = abc()
print(a)
#print(list(a))
print(next(a))
print(next(a))
print(next(a))
print(next(a))
print(next(a))
print(next(a))
print(list(a))
#print(next(a))
```
